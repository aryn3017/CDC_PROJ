```python
from torchvision import transforms

train_img_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
import os
import torch
from torch.utils.data import Dataset
from PIL import Image
import numpy as np

class PropertyDataset(Dataset):
    def __init__(
        self,
        df,
        targets,
        image_dir,
        tabular_features,
        scaler,
        img_transform=None,
        is_train=True
    ):
        self.df = df.reset_index(drop=True)
        self.targets = targets
        self.image_dir = image_dir
        self.tabular_features = tabular_features
        self.scaler = scaler
        self.img_transform = img_transform
        self.is_train = is_train

        # Scale tabular data
        self.tabular_data = self.scaler.transform(
            self.df[self.tabular_features]
        ).astype(np.float32)

        if self.is_train:
            self.targets = np.array(self.targets).astype(np.float32)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # Load image
        img_id = int(row['id'])
        img_path = os.path.join(self.image_dir, f"{img_id}.png")

        
        image = Image.open(img_path).convert("RGB")

        if self.img_transform:
            image = self.img_transform(image)

        tabular = torch.tensor(self.tabular_data[idx])

        if self.is_train:
            target = torch.tensor(self.targets[idx])
            return image, tabular, target
        else:
            return image, tabular
```

```python
train_dataset = PropertyDataset(
    df=train_inputs,
    targets=train_targets,
    image_dir="map_images/train",
    tabular_features= FEATURES_TO_SCALE,
    scaler=scaler,
    img_transform=train_img_transform,
    is_train=True
)

val_dataset = PropertyDataset(
    df=val_inputs,
    targets=val_targets,
    image_dir="map_images/train",
    tabular_features= FEATURES_TO_SCALE,
    scaler=scaler,
    img_transform=val_img_transform,
    is_train=True
)
```

```python
import torch
import torch.nn as nn

class MultimodalRegressor(nn.Module):
    def __init__(self, image_encoder, tabular_dim):
        super().__init__()

        self.image_encoder = image_encoder

        fusion_dim = 2048 + tabular_dim  # 2063

        self.regressor = nn.Sequential(
            nn.Linear(fusion_dim, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.3),

            nn.Linear(512, 128),
            nn.ReLU(),
            nn.BatchNorm1d(128),
            nn.Dropout(0.2),

            nn.Linear(128, 1)
        )

    def forward(self, image, tabular):
        img_features = self.image_encoder(image)
        x = torch.cat([img_features, tabular], dim=1)
        out = self.regressor(x)
        return out.squeeze(1)
```

```python
import torch
import numpy as np

def rmse_loss(y_pred, y_true):
    return torch.sqrt(torch.mean((y_pred - y_true) ** 2))
```

```python
def train_one_epoch(model, loader, optimizer, criterion, device):
    model.train()
    running_loss = 0.0

    for images, tabular, targets in loader:
        images = images.to(device)
        tabular = tabular.to(device)
        targets = targets.to(device)

        optimizer.zero_grad()

        outputs = model(images, tabular)
        loss = criterion(outputs, targets)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

    return running_loss / len(loader.dataset)
```

```python
def validate(model, loader, criterion, device):
    model.eval()
    preds, trues = [], []

    with torch.no_grad():
        for images, tabular, targets in loader:
            images = images.to(device)
            tabular = tabular.to(device)
            targets = targets.to(device)

            outputs = model(images, tabular)

            preds.append(outputs.cpu())
            trues.append(targets.cpu())

    preds = torch.cat(preds)
    trues = torch.cat(trues)

    rmse = rmse_loss(preds, trues)
    return rmse.item()
```

```python
import torch

class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.target_layer = target_layer
        self.activations = None
        self.gradients = None
        self._register_hooks()

    def _register_hooks(self):
        def forward_hook(module, input, output):
            self.activations = output

        def backward_hook(module, grad_input, grad_output):
            self.gradients = grad_output[0]

        self.target_layer.register_forward_hook(forward_hook)
        self.target_layer.register_full_backward_hook(backward_hook)

    def generate(self, image):
        """
        Grad-CAM for CNN feature learning (NOT final price).
        This is the correct approach when tabular features dominate.
        """
        self.model.zero_grad()

        # 🔥 Forward through CNN ONLY
        features = self.model.image_encoder(image)
        score = features.mean()     # scalar

        # 🔥 Backward from CNN activation
        score.backward()

        # Grad-CAM computation
        weights = self.gradients.mean(dim=(2, 3), keepdim=True)
        cam = (weights * self.activations).sum(dim=1)

        cam = torch.relu(cam)
        cam -= cam.min()
        cam /= (cam.max() + 1e-8)

        return cam.detach().cpu().numpy()
```

```python
from torch.utils.data import Dataset
from PIL import Image
import torch
import os

class TestMultimodalDataset(Dataset):
    def __init__(self, df, image_dir, tabular_features, scaler, transform=None):
        self.df = df.reset_index(drop=True)
        self.image_dir = image_dir
        self.tabular_features = tabular_features
        self.scaler = scaler
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        # image
        img_path = os.path.join(self.image_dir, f"{int(row['id'])}.png")
        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        # tabular
        tabular = self.scaler.transform(
            row[self.tabular_features].values.reshape(1, -1)
        )[0]
        tabular = torch.tensor(tabular, dtype=torch.float32)

        return image, tabular, int(row["id"])
```
