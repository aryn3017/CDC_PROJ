```python
# Compute correlation of numeric features with `price` and list them
import pandas as pd
import numpy as np

# train = pd.read_csv("data/train.csv")

numeric = train.select_dtypes(include=[np.number])
numeric = numeric.drop('price' , axis=1)
# Ensure 'price_log' exists
if 'price_log' not in numeric.columns:
    raise KeyError("'price_log' column is not numeric or not present in the dataset")

corr_with_price = numeric.corr()['price_log'].drop('price_log').sort_values(ascending=False)

print("Sorted by descending correlation (positive -> negative):")
print(corr_with_price.to_string())

print('\nSorted by descending absolute correlation (strength):')
print(corr_with_price.abs().sort_values(ascending=False).to_string())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=train
def plot_box(feature, data=df):
    data[feature+'_bin'] = pd.qcut(data[feature], q=20, duplicates='drop')
    plt.figure(figsize=(10, 5))
    sns.boxplot(
        x=data[feature+"_bin"],
        y=data["price_log"]
    )
    plt.xlabel(feature + " (Quantile Bins)")
    plt.ylabel("log(1 + Price)")
    plt.title(feature + " vs log(Price)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
features = ['sqft_living', 'sqft_above', 'sqft_living15', 'sqft_basement', 'sqft_lot', 'sqft_lot15', 'yr_built']

for feature in features:
    plot_box(feature)
```

```python
def plot_scatter(feature, data=df):
    plt.figure(figsize=(7, 5))
    sns.scatterplot(
        x=data[feature],
        y=data["price_log"],
        alpha=0.4,
        s=20
    )
    plt.xlabel(feature)
    plt.ylabel("log(1 + Price)")
    plt.title(f"{feature} vs log(Price)")
    plt.tight_layout()
    plt.show()

```

```python
from PIL import Image
import os
high_price_df = df.sort_values("price", ascending=False).head(3)
low_price_df  = df.sort_values("price", ascending=True).head(3)
```

```python
def show_images(df_subset, image_dir, title):
    plt.figure(figsize=(12, 4))

    for i, (_, row) in enumerate(df_subset.iterrows()):
        img_path = os.path.join(image_dir, f"{row['id']}.png")
        img = Image.open(img_path)

        plt.subplot(1, 3, i + 1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(f"ID: {row['id']}\nPrice: ${row['price']:,}")

    plt.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
show_images(
    low_price_df,
    image_dir="data/images/train",
    title="Low-Priced Properties (Bottom 3)"
)
```

```python
show_images(
    high_price_df,
    image_dir="data/images/train",
    title="High-Priced Properties (Top 3)"
)
```

```python
features = ['sqft_living', 'grade', 'sqft_above', 'sqft_living15', 'bathrooms', 'view', 'sqft_basement', 'bedrooms', 'floors', 'waterfront', 'yr_renovated', 'sqft_lot', 'sqft_lot15', 'yr_built', 'condition']

X = train[features]
y = train["price_log"]

print(X.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

model = RandomForestRegressor(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
preds = model.predict(X_val)

rmse = mean_squared_error(y_val, preds, squared=False)
r2   = r2_score(y_val, preds)

rmse, r2
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
xgb = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)
xgb.fit(X_train, y_train)
preds = xgb.predict(X_val)
rmse = mean_squared_error(y_val, preds)**0.5
r2   = r2_score(y_val, preds)
rmse, r2
```
